Part 0 — First, what even is a notebook?
A notebook is a document made of cells. You'll use two kinds:

Markdown cells hold formatted text (like this one). Double-click to edit, Shift+Enter to render.
Code cells hold Python you can actually run. The output appears right underneath.
That mix, runnable code sitting next to written explanation, is the whole point. It's why notebooks are the default tool for exploring data: you poke at something, see the result immediately, write a note, poke again.

The kernel remembers things between cells
Behind the notebook is one live Python process called the kernel. It stays running, so a variable you create in one cell is still around in the next one. Run these two cells in order and you'll see the second one happily uses what the first one defined.

Run order matters (the classic gotcha)
Cells run in the order you run them, not necessarily top to bottom. The [ ] to the left of a code cell shows a number once it's run, and that number is the order it ran in. If your notebook starts behaving strangely (a variable "vanishes", or has an old value), it's almost always run-order confusion.

💡 The fix, and the best habit you can build: Kernel → Restart Kernel and Run All Cells. It wipes the slate and runs everything fresh, top to bottom, the way another person (or future you) would.

Notebooks + Git: clear your outputs before you commit
A notebook saves its outputs inside the file. That's lovely for reading, but rough for Git: every re-run changes the saved output, so your diffs get noisy and the repo bloats, sometimes with huge image or data blobs.

The fix is a one-time habit before each commit:

⚠️ Kernel → Restart Kernel and Clear All Outputs (or Edit → Clear All Outputs), then git add and git commit.

That keeps your version history about the code you wrote, not the output noise. Every commit prompt in this challenge assumes you've cleared outputs first.

Stage 2 — Make it a record, add a function and a decision
Lesson. A dictionary groups related fields under labels, much tidier than five loose variables. A function names a piece of reusable work. if/elif/else lets your code decide something.

Task.

Put the book into a dict: book = {"title": ..., "pages": ..., "minutes": ..., "genre": ...}.
Write pace(book) that returns pages per minute (pages / minutes).
Write speed_label(book) that returns "fast", "steady" or "slow" using if/elif/else on the pace (suggested: > 0.8 fast, > 0.6 steady, else slow).
Print the label for your book.
💾 Commit: git commit -m "Stage 2: book record, pace() and speed_label()"


# --- YOUR CODE (Stage 2) ---




```python
book = {
    "title": "The Hobbit",
    "pages": 310,
    "minutes": 400,
    "genre": "Fantasy"
}

# Function to calculate reading pace
def pace(book):
    return book["pages"] / book["minutes"]

# Function to classify reading speed
def speed_label(book):
    reading_pace = pace(book)

    if reading_pace > 0.8:
        return "fast"
    elif reading_pace > 0.6:
        return "steady"
    else:
        return "slow"

# Print the results
print(f"Book: {book['title']}")
print(f"Genre: {book['genre']}")
print(f"Reading pace: {pace(book):.2f} pages/minute")
print(f"Speed label: {speed_label(book)}")
```

```python
# ---------- Stage 3: Many books (a list) and a loop ----------
books = [
    {"title": "Dune",                 "pages": 412, "minutes": 540, "genre": "sci-fi"},
    {"title": "The Hobbit",           "pages": 310, "minutes": 360, "genre": "fantasy"},
    {"title": "Sapiens",              "pages": 498, "minutes": 720, "genre": "non-fiction"},
    {"title": "Project Hail Mary",    "pages": 476, "minutes": 600, "genre": "sci-fi"},
    {"title": "The Name of the Wind", "pages": 662, "minutes": 840, "genre": "fantasy"},
    {"title": "Educated",             "pages": 334, "minutes": 480, "genre": "non-fiction"},
]

total_pages = 0

print("Books:\n")

for book in books:
    print(f"{book['title']} - {book['pages']} pages")
    total_pages += book["pages"]

print("\n-----------------------")
print(f"Total pages: {total_pages}")
print(f"Number of books: {len(books)}")

```
